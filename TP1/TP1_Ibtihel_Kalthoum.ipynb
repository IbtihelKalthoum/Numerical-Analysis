## Fonction horner

```python
def horner(P,x):
    n=len(P)
    val=P[n-1]
    for i in range(n-2,-1,-1):
        val=val*x+P[i]
    return val
```

```python
import numpy as np
T=[1,-6, 15, -20, 15, -6, 1]
print('Expression réduite')
print('***p(i)***')
for i in np.arange(0.998, 1.002,0.0001):
    print('p(i)= ' , p(i))

print('Expression developpée')
print('***f(i)***')
for i in np.arange(0.998, 1.002,0.0001):
    print('f(i)= ' , f(i))
print('***horner***')
for i in np.arange(0.998, 1.002,0.0001):
    print('horner(T,i) =' , horner(T,i))
```

### Donner la version python de `dicho`

```python
def dicho(a,b,f,eps):
    mil=(a+b)/2
    ecar=b-a
    while ecar > eps :
        if f(mil)==0:
            break
        
        if f(a)*f(mil)<0:
            b=mil
        else :
            a=mil
            
            
            return (a,b),mil
        f=lambda x: x**3+x**2-3*x-3
```

### 3)

Le nombre d'itération N nécéssaire à la méthode de la dichotomie pour trouver $\alpha$ à $10^{-p}$ prés 
                         $$N \geq E\left(\dfrac{\ln(b-a)+p\ln(10)}{\ln(2)}\right)+1$$

```python
def dichow(a,b,f,eps):
    
    n=0
    while (b-a)>eps:
        m=(a+b)/2
        if f(m)==0:
            return m
        if f(a)*f(m)<0 :
            b=m
        else :
            a=m
        n+=1
    return m,n
```

```python
interval=[[-2,-1.5],[-1.5,-0.5],[1.5,2]]
for u in interval:
    print((dichow(u[0],u[1],f,0.001)))
```

## Conclusion :
**Les Avantages:**

la méthode de la dichotomie est une méthode :

-simple

-les nombres d'itérations est calculé facilement si on connait epsilon

-Le principal avantage pratique de cette méthode est sa robustesse, puisque si f est continue, alors l'algorithme est théoriquement convergent (la taille de l'intervalle de recherche tend vers zéro)

**Les inconvénients :**

Le principal défaut de l'algorithme est que seul le signe de f est utilisé, ce qui mène à une convergence plutôt lente (convergence quasiment linéaire).

donc on va chercher une autre méthode plus rapide par exemple la méthode de newton ou la méthode de point fixe

>1) On considère la fonction $f(x)=e^{-x}-x$. Ecrire puis appliquer  une fonction Matlab que l'on appellera `newton.m` à la >fonction $f$, en prenant $x0=0$, $eps=0.001$ .

##  la solution entière avec un code python

### Solution Python
#### 1)

```python
from math import exp
def newtonn(a,eps):
    x1=a
    x2=x1-(exp(-x1)-x1)/(-1*exp(-x1)-1)
    while abs(x1-x2)>eps:
        x1=x2
        x2=x1-(exp(-x1)-x1)/(-1*exp(-x1)-1)
    return x2

```
